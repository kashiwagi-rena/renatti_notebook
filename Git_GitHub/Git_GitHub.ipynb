## カリキュラムに出てきたコマンドの中でわからなかったコマンド集
1. ブランチ作らずに、mainに追記しちゃった〜〜〜<br>
    →スタッシュしましょう！<br>
    
    ```jsx
     $ git stash -u
     # 全ての変更ファイルを退避
     
     $ git stash list
     # 退避させたもののリスト確認
     
     $ git stash apply stash@{0}
     # 退避させた変更ファイルを戻す
    ```
    [【git stash】コミットはせずに変更を退避したいとき ： Qiita](https://qiita.com/chihiro/items/f373873d5c2dfbd03250)
    
2. git add .を多用しない理由<br>
    私が思うになのですが、ファイルの変更点をちゃんと理解せずに全て入れてしまうのってかなり危険だからだと思います。<br>
    例えば、誤ってアクセスキーやAWS系の記述をそもままGitHubあげてしまうことがあるかもしれません。それを防ぐためにファイル名を指定しているのではないでしょうか？<br>

## PRやマージって、何をするのかわかりますか？
PR→人にコードの記述内容を確認とる際に使用するコミニュケーションツール<br>
pull request の作成 - GitHub Docs：https://docs.github.com/ja/pull-requests/collaborating-with-pull-requests/proposing-changes-to-your-work-with-pull-requests/creating-a-pull-request<br>
<br>
マージ→ブランチを別のブランチに統合すること<br>
<br>
<br>
## PRとは？
ブランチ一つに対して、一つ作成可能<br>
他の人に確認を取るためのシステム<br>
<br>
作り方<br>
1. 新しいブランチを作成<br>
    
    ```jsx
    % git checkout -b feature/ブランチ名
    ```
    
2. ファイル変更<br>
3. コミットする<br>
    
    ```jsx
    % git add .
    
    % git commit -m "コミットメッセージ"
    ```
    
4. pushする<br>
    
    ```jsx
    % git push origin feature/ブランチ名
    ```
    
5. GitHub上の該当リポジトリに飛ぶ→「new pull request」を押下<br>
